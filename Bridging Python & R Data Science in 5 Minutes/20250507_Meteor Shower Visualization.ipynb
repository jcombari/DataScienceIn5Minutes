```r
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Simulate a dataset similar to R's quakes
np.random.seed(42)
data = pd.DataFrame({
    "lat": np.random.uniform(-38.5, -17.5, 1000),  # latitude range
    "long": np.random.uniform(165, 190, 1000),     # longitude range
    "depth": np.random.uniform(0, 700, 1000)       # depth range
})

# Create scatter plot using longitude and latitude
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data["long"], data["lat"], c=data["depth"], cmap='viridis', s=data["depth"] / 2, alpha=0.6)
plt.colorbar(scatter, label='Depth (km)')  # Add color bar for depth
plt.title("Simulated Meteor Shower (Earthquake Data)")  # Title for the plot
plt.xlabel("Longitude")  # X-axis label
plt.ylabel("Latitude")  # Y-axis label
plt.grid(True)  # Show grid for better readability
plt.show()  # Display the plot

```

```r
# Load necessary libraries
suppressWarnings(library(ggplot2))

# Load the built-in 'quakes' dataset
data("quakes")

# Create a scatter plot: longitude vs latitude, point size by depth
ggplot(quakes, aes(x = long, y = lat)) +
  geom_point(aes(size = depth, color = depth), alpha = 0.6) +  # Set point size and color by depth
  scale_color_viridis_c() +  # Use a color scale (viridis)
  labs(
    title = "Meteor Shower Simulation with Earthquake Data",  # Title of the plot
    x = "Longitude",  # X-axis label
    y = "Latitude",   # Y-axis label
    color = "Depth (km)",  # Color label for depth
    size = "Depth (km)"    # Size label for depth
  ) +
  theme_minimal()  # Use a minimal theme for a clean look
```
