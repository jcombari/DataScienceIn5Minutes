```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Cargar el dataset de MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255.0

# Convertir las etiquetas a formato categ√≥rico (one-hot encoding)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Definir el modelo de red neuronal convolucional aleatoria con refuerzo
class RRCN(tf.keras.Model):
    def __init__(self):
        super(RRCN, self).__init__()
        # Generar filtros aleatorios para la capa convolucional
        self.random_filter = tf.Variable(np.random.randn(3, 3, 1, 32), dtype=tf.float32)  # Inicializaci√≥n aleatoria de filtros
        self.conv_layer = Conv2D(32, (3, 3), padding='same', activation='relu')
        self.flatten = Flatten()
        self.fc_layer = Dense(128, activation='relu')
        self.output_layer = Dense(10, activation='softmax')
    
    def call(self, inputs):
        # Aplicar filtros aleatorios
        x = self.conv_layer(inputs)
        x = self.flatten(x)
        x = self.fc_layer(x)
        return self.output_layer(x)

    def update_filters(self, reward):
        """
        M√©todo para actualizar los filtros de forma sencilla.
        Si la recompensa es positiva, se amplifican los valores de los filtros, 
        si la recompensa es negativa, se ajustan para hacerlos m√°s aleatorios.
        """
        if reward > 0:
            self.random_filter.assign(self.random_filter * 1.1)  # Refuerzo positivo (amplificando filtros)
        else:
            self.random_filter.assign(self.random_filter * 0.9)  # Refuerzo negativo (ajustando filtros hacia valores m√°s peque√±os)

# Crear el modelo RRCN
model = RRCN()

# Compilar el modelo con optimizador Adam
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo con refuerzo
for epoch in range(5):  # Entrenamos por 5 √©pocas
    # Entrenamos en lotes
    history = model.fit(X_train, y_train, epochs=1, batch_size=32, validation_data=(X_test, y_test))
    
    # Evaluamos el modelo despu√©s de cada √©poca
    val_loss, val_acc = model.evaluate(X_test, y_test)
    
    # Imprimir la precisi√≥n de la validaci√≥n
    print(f"Epoch {epoch + 1} - Validation accuracy: {val_acc}")
    
    # Asignar una recompensa basada en la precisi√≥n
    if val_acc > 0.98:  # Si la precisi√≥n de validaci√≥n es mayor al 98%, recompensa positiva
        reward = 1
    else:  # Si la precisi√≥n es baja, recompensa negativa
        reward = -1
    
    # Actualizar los filtros con base en la recompensa
    model.update_filters(reward)

# Evaluar el modelo al final
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Final Test accuracy: {test_acc}")

# Graficar la precisi√≥n durante el entrenamiento
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```
