- 📌 **Reinforcement Learning**: This technique optimizes the system's decision-making by learning from past interactions and selecting the most relevant features to predict the best matches.

By combining these two approaches, the **RRCN** model can provide more personalized recommendations, helping platforms better predict mutual preferences, going beyond the limitations of traditional methods.

## 🔧 **¿Cómo funciona?**

El modelo **RRCN (Reinforced Random Convolutional Network)** combina dos poderosas técnicas de inteligencia artificial para mejorar las recomendaciones recíprocas en plataformas de citas online:

- 📌 **Redes Neuronales Convolucionales (CNN)**: Estas redes son ideales para identificar patrones en datos complejos, como los intereses o preferencias de los usuarios.
- 📌 **Aprendizaje por Refuerzo (Reinforcement Learning)**: Este enfoque optimiza las decisiones del sistema al aprender de las interacciones previas y seleccionar las características más relevantes para predecir las mejores coincidencias.

Al utilizar estas dos técnicas, el modelo **RRCN** es capaz de hacer recomendaciones más personalizadas, ayudando a las plataformas a predecir con mayor precisión las preferencias mutuas de los usuarios, superando las limitaciones de los métodos tradicionales.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Cargar el dataset de MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255.0

# Convertir las etiquetas a formato categórico (one-hot encoding)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Definir el modelo de red neuronal convolucional aleatoria con refuerzo
class RRCN(tf.keras.Model):
    def __init__(self):
        super(RRCN, self).__init__()
        # Generar filtros aleatorios para la capa convolucional
        self.random_filter = tf.Variable(np.random.randn(3, 3, 1, 32), dtype=tf.float32)  # Inicialización aleatoria de filtros
        self.conv_layer = Conv2D(32, (3, 3), padding='same', activation='relu')
        self.flatten = Flatten()
        self.fc_layer = Dense(128, activation='relu')
        self.output_layer = Dense(10, activation='softmax')
    
    def call(self, inputs):
        # Aplicar filtros aleatorios
        x = self.conv_layer(inputs)
        x = self.flatten(x)
        x = self.fc_layer(x)
        return self.output_layer(x)

    def update_filters(self, reward):
        """
        Método para actualizar los filtros de forma sencilla.
        Si la recompensa es positiva, se amplifican los valores de los filtros, 
        si la recompensa es negativa, se ajustan para hacerlos más aleatorios.
        """
        if reward > 0:
            self.random_filter.assign(self.random_filter * 1.1)  # Refuerzo positivo (amplificando filtros)
        else:
            self.random_filter.assign(self.random_filter * 0.9)  # Refuerzo negativo (ajustando filtros hacia valores más pequeños)

# Crear el modelo RRCN
model = RRCN()

# Compilar el modelo con optimizador Adam
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo con refuerzo
for epoch in range(5):  # Entrenamos por 5 épocas
    # Entrenamos en lotes
    history = model.fit(X_train, y_train, epochs=1, batch_size=32, validation_data=(X_test, y_test))
    
    # Evaluamos el modelo después de cada época
    val_loss, val_acc = model.evaluate(X_test, y_test)
    
    # Imprimir la precisión de la validación
    print(f"Epoch {epoch + 1} - Validation accuracy: {val_acc}")
    
    # Asignar una recompensa basada en la precisión
    if val_acc > 0.98:  # Si la precisión de validación es mayor al 98%, recompensa positiva
        reward = 1
    else:  # Si la precisión es baja, recompensa negativa
        reward = -1
    
    # Actualizar los filtros con base en la recompensa
    model.update_filters(reward)

# Evaluar el modelo al final
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Final Test accuracy: {test_acc}")

# Graficar la precisión durante el entrenamiento
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```
