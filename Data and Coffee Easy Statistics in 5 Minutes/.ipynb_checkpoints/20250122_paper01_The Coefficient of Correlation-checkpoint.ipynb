### ‚ùì How do we know if two variables are related and to what extent?
üôÉ With Pearson's Correlation Coefficient!

### üîß How does it work?
Pearson's coefficient evaluates:
- üåü **Strength**: How closely related are the variables?
- üåü **Direction**: Is the relationship positive or negative?

It‚Äôs a value between -1 and 1: 
- üìâ **-1**: Perfect negative linear relationship (one increases as the other decreases).
- üìà **1**: Perfect positive linear relationship (both increase together).
- ‚ûñ **0**: No linear relationship.

### üîé Why does it matter?
Knowing this relationship allows you to make informed decisions, such as:
- üìà Predicting behaviors, like how much tip customers might leave.
- üìâ Identifying anomalies, such as high bills with low tips.

### ‚ú® Common mistakes when using it:
- ‚ö†Ô∏è Assuming causation: Correlation does not imply causation!
- ‚ö†Ô∏è Using categorical data or mixed data: It only works with quantitative variables.

### ‚ú® Practical example:
- üìä **Dataset**: tips from seaborn. This dataset includes real restaurant data:
    - total_bill: Total bill amount.
    - tip: Tip amount.
    - sex: Gender of the customer.
    - day: Day of the week.

- ‚öôÔ∏è **Business impact**:
This analysis could help you tailor your service strategies by identifying patterns: Do customers tip more on weekends? Do higher bills lead to larger tips?

### üìä Code summary:
Using Python, we:
- Visualize the relationship between total_bill and tip with a scatterplot.
- Calculate Pearson's correlation coefficient.
- Interpret the result in terms of the relationship's strength and direction.

### ‚ú¶‚úß What do you get with this code?
The correlation coefficient is **0.67**, indicating:
- A positive and moderately strong relationship between the bill and the tip.
- This means that as the total bill increases, the tip tends to increase as well.

# üêç Python code

üìñ‚úçÔ∏è Paper name: "The Coefficient of Correlation"
Author: Karl Pearson
Year: 1909

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the 'tips' dataset from seaborn
data = sns.load_dataset('tips')

# Scatterplot to visualize the relationship between 'total_bill' and 'tip'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='total_bill', y='tip', alpha=0.7, color='blue')
plt.title('Scatterplot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.grid(True)
plt.show()

# Calculate Pearson's correlation coefficient
correlation, p_value = pearsonr(data['total_bill'], data['tip'])

# Display the results
print(f"Pearson's Correlation Coefficient: {correlation:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation:
# A correlation coefficient of ~0.67 indicates a positive, moderately strong linear relationship.
# P-value < 0.05 suggests the relationship is statistically significant.
```
